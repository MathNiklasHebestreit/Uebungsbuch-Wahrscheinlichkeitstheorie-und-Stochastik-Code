**(Aufgabe 78, Natriumkonzentration in Leitungswasser).**
Aufgrund langjähriger Untersuchungen ist bekannt, dass die Natriumkonzentration in einem Liter Leitungswasser normal-verteilt mit Erwartungswert $\mu = 42 \operatorname{mg}$ und Standardabweichung $\sigma = 5 \operatorname{mg}$ ist. Lösen Sie die folgenden Teilaufgaben mit $\texttt{Python}$:

(a) Skizzieren Sie die Wahrscheinlichkeit, dass die Natriumkonzentration zwischen $37 \operatorname{mg}$ und $52 \operatorname{mg}$ liegt und bestimmen Sie diese anschließend.

(b) Bestimmen Sie die Wahrscheinlichkeit, dass die Natriumkonzentration höchs"-tens $32 \operatorname{mg}$ beträgt.

(c) Untersuchen Sie, welche Natriumkonzentration mit einer Wahrscheinlichkeit von $97.5 \%$ überschritten wird.

**Lösung.**

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters of the normal distribution
mu = 42   # Mean sodium concentration in mg
sigma = 5 # Standard deviation in mg
```

**Lösung (a).**

```python
# Part (a). Probability that sodium concentration lies between 37 mg and 52 mg

a, b = 37, 52
p = norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma)

# Plot the PDF with shaded area between a and b
plt.figure(figsize=(10, 5))
plt.plot(x, pdf, label='Normal distribution')
plt.fill_between(x, pdf, where=(x >= a) & (x <= b), alpha=0.3, label=f"P({a} ≤ X ≤ {b}) = {p:.4f}")
plt.title('Normal Distribution of Sodium Concentration in Water')
plt.xlabel('Sodium concentration')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(f"The probability that sodium concentration lies between 37 mg and 52 mg is {p * 100:.2f} %.")
```

**Lösung (b).**

```python
# Part (b). Probability that sodium concentration is at most 32 mg

p = norm.cdf(32, mu, sigma)
print(f"The probability that sodium concentration is at most 32 mg is {p * 100:.2f} %.")
```

**Lösung (c).**

```python
# Part (c). Value exceeded with 97.5% probability

p = norm.ppf(1 - 0.975, mu, sigma)  # Inverse CDF for lower 2.5%
print(f"With 97.5% probability, the sodium concentration exceeds at least {p:.2f} mg.")
```
